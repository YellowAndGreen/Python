```python
from selenium import webdriver
import time
driver = webdriver.Edge()
driver.get('https://eniu.com/gu/sz000002#')

button = driver.find_element_by_xpath("/html/body/div[4]/div/div[5]/div/div[4]/div/div[2]/div/div[2]/div/div[1]/h4/a")
button.click()

pbpe = driver.find_element_by_xpath("/html/body/div[4]/div/div[5]/div/div[4]/div/div[2]/div/div[2]/div/div[2]/div/table/tbody")
text = pbpe.get_attribute('textContent')
driver.close()
driver.quit()
```

```python
def text_parser(text):
    year,pe_min,pe_max,pe_avg=[],[],[],[]
    a = text.split()
    b = [a[i:i+4] for i in range(0,len(a),4)] # 按四个一组分组
    for i in b:
        year.append(i[0])
        pe_avg.append(i[1])
        pe_max.append(i[2])
        pe_min.append(i[3])
    return year,pe_avg,pe_max,pe_min
```
