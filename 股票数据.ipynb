```python
def cal(code_val='sh.600048'):
    df = pd.read_excel("E:\\All Files\\stock\\"+code_val+".xls","Sheet1")
    date= []
    for i in df.Date.to_list():
        date.append(str(i.year)+"/"+str(i.month)+"/"+str(i.day))
    data = []
    for i in range(len(date)):
        data.append([date[i],df.Open.to_list()[i],df.Close.to_list()[i],df.Low.to_list()[i],df.High.to_list()[i]])
    print(data)
cal(code_val = sys.argv[1])
```

## 盘面

```python
import baostock as bs
import pandas as pd

def find_stock_code():
    #### 登陆系统 ####
    lg = bs.login()

    #### 获取证券信息 ####
    rs = bs.query_all_stock(day="2021-12-31")

    #### 打印结果集 ####
    data_list = []
    while (rs.error_code == '0') & rs.next():
        # 获取一条记录，将记录合并在一起
        data_list.append(rs.get_row_data())
    result = pd.DataFrame(data_list, columns=rs.fields)
    stock_code = result["code"]

    #### 结果集输出到csv文件 ####   
    result.to_csv("E:\\All Files\\stock\\index\\index.csv", encoding="gbk", index=False)

    #### 登出系统 ####
    bs.logout()
    return stock_code
```

```python
def get_data():
    #### 登陆系统 ####

    #### 所有代码 ####
    stock_code = find_stock_code()
    progress = len(stock_code)
    for index in range(progress):
        lg = bs.login()
        #### 获取沪深A股历史K线数据 ####
        # adjustflag：复权类型，默认不复权：3；1：后复权；2：前复权
        code = stock_code[index]
        rs = bs.query_history_k_data_plus(code,
        "date,code,open,high,low,close,preclose,volume,amount,adjustflag,turn,tradestatus,pctChg,peTTM, pbMRQ, psTTM, pcfNcfTTM, isST",
        start_date='2019-01-01', end_date='2021-12-31',
        frequency="d", adjustflag="1")
        #### 打印结果集 ####
        data_list = []
        while (rs.error_code == '0') & rs.next():
            # 获取一条记录，将记录合并在一起
            data_list.append(rs.get_row_data())
        result = pd.DataFrame(data_list, columns=rs.fields)

        #### 结果集输出到csv文件 ####   
        result.to_csv("E:\\All Files\\stock\\index\\后复权\\"+code+".csv", index=False)
        print(str(index)+"/"+str(progress))
        bs.logout()
```
