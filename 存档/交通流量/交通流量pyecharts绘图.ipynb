```python
import numpy as np,math,pandas as pd
from pyecharts.charts import *
from pyecharts import options as opts
from pyecharts.charts import Line
from pyecharts.globals import ThemeType
bar1 = (Bar(init_opts=opts.InitOpts(width='800px', height="400px"))
    .add_xaxis(data1["Unnamed: 0"].tolist())
    .add_yaxis('3车道流量', data1["3车道流量"].tolist())
    .add_yaxis('4车道流量', data1["4车道流量"].tolist())
    .set_global_opts(
    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=35)),
    yaxis_opts=opts.AxisOpts(name="流量数"),
    )

    .set_series_opts(label_opts=opts.LabelOpts(interval=0))       
       )


bar1.render_notebook()
```

```python
bar1 = (Bar(init_opts=opts.InitOpts(width='800px', height='400px'))
    .add_xaxis(data2["Unnamed: 0"].tolist())
    .add_yaxis('3车道频数', data2["3车道频数"].tolist())
    .set_global_opts(
    yaxis_opts=opts.AxisOpts(name="频数"

)))
line1 = (Line(init_opts=opts.InitOpts(width='1200px', height='400px'))
.add_xaxis(data2["Unnamed: 0"].tolist())
.add_yaxis('3车道累计频率',[21.78, 37.78, 52.09, 70.73, 85.82, 94.97, 100.0],z=100.0)
.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
.set_global_opts(
legend_opts=opts.LegendOpts(is_show=True,
                                                 
                                                 pos_bottom='90%'),
    yaxis_opts=opts.AxisOpts( position="right",splitline_opts=opts.SplitLineOpts(is_show=True),name="累积频率%"

)))

grid = (Grid(init_opts=opts.InitOpts(width='800px', height='400px',theme="vintage"))
        .add(bar1,grid_opts=opts.GridOpts(pos_left="15%")) 
        .add(line1,grid_opts=opts.GridOpts(pos_left="15%"))
        )
grid.render("3.html")
```
